### ROS Package
Your package will depend on rospy, roscpp, std_msgs, geometry_msgs and duckietown_msgs. 

You can name the package as you wish.

### ROS Nodes
You will create one node in your package.
A template is provided.

**Square Node**: 
This node will be similar to the your Ex2 node. 
It should make the MM move in a square.
The node should perform the following:
- Drive straight for 1m then stop for 5 seconds.  
DEMONSTRATE this before moving on.
- Turn 90 degrees, drive 1m then stop for 5 seconds.  
DEMONSTRATE this before moving on.
- Turn 90 degrees
- Repeat THESE EXACT COMMANDS twice more to make something like a square.  

Note that in the real MM, it **will not be a perfect square**.   
However, in the MM simulator, it **must be a perfect square**.  
For the real MM, only the first two corners are graded.  
DEMONSTRATE this.

- **Note**- ROS nodes will start at different times, so consider adding a slight delay before publishing commands.
Add a delay of at least 10 seconds in the beginning of your node python script.  
The simulator takes some time to start, the delay prevents you from starting your node before the simulator is ready.
See the template.


### ROS Launch file
The launch file must start everything needed to accomplish the exercise objectives.
Your launch file must start
1. Your Square Node
2. Rqt graph
3. Rostopic echo MM pose topic output
